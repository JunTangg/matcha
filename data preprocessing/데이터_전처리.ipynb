# 데이터 전처리

```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv('네이버맛집_이미지추가.csv', header='infer', engine='python', encoding='utf-8-sig', sep=',')
```

```python
df['Menu'][24]
```

```python
# Address열 전처리
for i, df_Address in enumerate(df.Address):
    df.loc[i, 'Address'] = re.sub('지도보기', '', df_Address)
```

```python
# Menu 열 전처리1
for i, df_Menu in enumerate(df.Menu):
    df.loc[i, 'Menu'] = re.sub('원\r\n', '원 ', df_Menu)
```

```python
# Menu 열 전처리2
for i, df_Menu in enumerate(df.Menu):
    df.loc[i, 'Menu'] = re.sub('대표\r\n', '', df_Menu)
```

```python
# Menu 열 전처리3
for i, df_Menu in enumerate(df.Menu):
    df.loc[i, 'Menu'] = re.sub('\r\n더보기', '', df_Menu)
```

```python
# Menu 열 전처리4
for i, df_Menu in enumerate(df.Menu):
    df.loc[i, 'Menu'] = re.sub('\r\n메뉴판.*', '', df_Menu)
```

```python
# Menu 열 전처리5
for i, df_Menu in enumerate(df.Menu):
    df.loc[i, 'Menu'] = re.sub('메뉴판.*', '', df_Menu)
```

```python
# Menu 열 전처리6
for i, df_Menu in enumerate(df.Menu):
    df.loc[i, 'Menu'] = re.sub('메뉴.*\r\n', '', df_Menu)
```

```python
# age 열 전처리
ages = ['10', '20', '30', '40', '50', '60']
for age in ages:    
    for i, df_prefer_per_age in enumerate(df[age]):
        try:
            df.loc[i, age] = re.sub('%;', '' , re.sub('height: ', '',df_prefer_per_age))
        except:
            df.loc[i, age] = ''
```

```python
# 성별열 전처리
genders = ['Female', 'Male']
for gender in genders:
    for i, df_prefer_per_gender in enumerate(df[gender]):
        try:
            df.loc[i, gender] = re.sub('%', '', df_prefer_per_gender)
        except:
            df.loc[i, gender] = ''
```

```python
# Store_name에 Category와 중복된 부분 제거
for i, df_Store_name in enumerate(df.Store_name):
    if str(df.loc[i, 'Category']) != 'nan':
        df.loc[i, 'Store_name'] = re.sub(df.loc[i, 'Category'], '', df_Store_name)
```

```python
# 바나프레소 오류 수정
for i, df_Category in enumerate(df.Category):
    if re.search(str(df_Category), '바나프레소'):
        df.loc[i, 'Category'] = re.sub('바나프레소', '카페', df_Category)
        df.loc[i, 'Store_name'] = '바나프레소' + df.loc[i, 'Store_name']
```

```python
# Review 컬럼 , 제거
for i, df_Review in enumerate(df.Review):
    df.loc[i, 'Review'] = re.sub(',', '', df_Review)
```

```python
df.to_csv("네이버맛집_전처리.csv", index=False, encoding='utf-8-sig')
```
